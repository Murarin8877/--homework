```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download stock data
data = yf.download('2330.TW', start='2024-10-01', end='2024-11-29')

# Calculate 10-day and 20-day moving averages
data['MA10'] = data['Close'].rolling(window=10).mean()
data['MA20'] = data['Close'].rolling(window=20).mean()

# Plot K-line chart
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', linestyle='-', marker='o', alpha=0.7)
plt.plot(data.index, data['MA10'], label='10-day MA', linestyle='--')
plt.plot(data.index, data['MA20'], label='20-day MA', linestyle='-.')

# Annotate chart
plt.title('2330.TW K-line Chart with MA10 & MA20', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (TWD)', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

# Save plot as an image
plt.tight_layout()
#chart_path = '/mnt/data/2330_K_Line_Chart.png'
#plt.show()
#儲存圖片
# Save plot to Colab environment
chart_path = '2330_K_Line_Chart.png'
plt.savefig(chart_path)
plt.show()
plt.close()

# Provide a link to download the image
from google.colab import files
img = files.download(chart_path)
```
